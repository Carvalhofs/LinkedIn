```python
#Importando bibliotecas que iremos utilizar.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Setando configuraes de dataframe para mostrar 80 colunar e 80 linhas.
pd.set_option('display.max_columns', 80 )
pd.set_option('display.max_rows', 80)
```

```python
#Atribuindo base de dados para varivel df_enem_bahia
df_enem_bahia = pd.read_csv("C:\Arquivos Felipe\Ciencia de dados - Digital House\case enem sql\enem_bahia.csv")
```

```python
df_enem_bahia
```

```python
#Observando quais variveis no so numricas, para decidir se sero removidas ou transformadas em numricas.
df_enem_bahia.select_dtypes(include='object')
```

```python
#Para essa anlise, vou desconsiderar as variveis que se referem a localidade ou a gabarito. Portanto irei utilizar uma estratgia simples para transformar 
#os dados das colunas Object de interesse em numricos.

df_enem_bahia.TP_SEXO = df_enem_bahia.TP_SEXO.replace({'M':0, 'F':1})
valores = {'':99,'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'H':8, 'I':9, 'J':10, 'K':11, 'L':12, 'M':13, 'N':14, 'O':15, 'P':16, 'Q':17, 'R':18, 'S':19, 'T':20}
for i in ['Q0'+'{:02}'.format(i) for i in range(1,26)]:
  df_enem_bahia[i] = df_enem_bahia[i].replace(valores)
```

```python
df_enem_bahia
```

```python
#Criando uma nova coluna chamada media_notas, que vai receber a media das 5 notas do enem por inscrito
df_enem_bahia['media_notas'] = (df_enem_bahia['NU_NOTA_CH'] + df_enem_bahia['NU_NOTA_CN'] + df_enem_bahia['NU_NOTA_MT'] + df_enem_bahia['NU_NOTA_LC'] + df_enem_bahia['NU_NOTA_REDACAO']) / 5
```

```python
df_enem_bahia
```

```python
#Avaliando a correlao entre as demais variveis e a coluna media_notas, selecionando apenas aquelas com correlao superior a 20%
correlacao = df_enem_bahia.corr()
correlacao = correlacao [correlacao['media_notas'].abs() >= .20]
correlacao
```

```python
#Algumas variveis possuem alta correlao com a renda familiar, e talvez fosse interessantes desconsider-las.
#No entrarei nesse tpico e manterei todas pois o Objetivo  meramente demonstrar a utilizao de algumas ferramentas e entregar um modelo em operao.
print('Correlao das variveis com a Renda Familiar')
print(correlacao['Q006'])
```

```python
#TP_DEPENDENCIA_ADM_ESC ser desconsiderada por ter muitas colunas NaN equivalente a 98% dos dados do DataFrame
#Plotando soma da quantidade de valores nicos para cada varivel.
df_enem_bahia.isna().sum()
```

```python
#Removendo valores NaN das colunas que sero usadas no modelo.
df_enem_bahia = df_enem_bahia.dropna(subset = ['media_notas','Q001', 'Q002', 'Q003', 'Q004',
       'Q006', 'Q008', 'Q013','Q014', 'Q016', 'Q018', 'Q019', 'Q021', 'Q022',
       'Q024', 'Q025'])
df_enem_bahia
```

```python
#Visualizando todas as colunas do DataFrame
df_enem_bahia.columns
```

```python
#Atribuinudo as colunas que sero "features"  varivel "X" e a coluna "target"  varivel "y"

X = df_enem_bahia[['Q001', 'Q002', 'Q003', 'Q004',
       'Q006', 'Q008', 'Q010', 'Q013','Q014', 'Q016', 'Q018', 'Q019', 'Q021', 'Q022',
       'Q024', 'Q025']]

y = df_enem_bahia['media_notas']
```

```python
#Importando ferramenta de separao de dados de treino e teste do Scikit Learn.
from sklearn.model_selection import train_test_split
```

```python
#Separando variveis de treino e teste para X e y
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
#Importando stats model

import statsmodels.api as sm
ols_regressor = sm.OLS(y_train, X_train)
model = ols_regressor.fit()
```

```python
#Verificando aspectos estatsticos do modelo.
#Teste de Significncia individuais ou p-values dos coeficientes: diz o quanto das variveis preditoras explicam a varivel alvo. A mtrica padro  o p-value ser menor que 0.05.
#Coeficiente R: diz o quanto o meu modelo explica seus resultados.  um valor entre 0 e 1. Quanto mais prximo de 1, melhor.

model.summary()
```

```python
#Criando varivel de previso utilizando modelo anterior.
pred_X_test = model.predict(X_test)
```

```python
#Criando novo DataFrame para receber resultados previstos pelo modelo e resultados reais separados para teste.
novo_df_enem_bahia = pd.DataFrame([pred_X_test, y_test])
novo_df_enem_bahia = novo_df_enem_bahia.transpose()
novo_df_enem_bahia.columns = ['pred_X_test', 'y_test']
novo_df_enem_bahia = novo_df_enem_bahia.reset_index().drop(columns = ['index'])
print(novo_df_enem_bahia)
```

```python
#Importando ferramentas de avalio de erro em modelo de regresso por mtodos estatsticos
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
#Definindo uma funo para tornar mais fcil a plotagem dos mtodos estatsticos quando necessrio utilizar
def desvio_3 (previsao, teste):
    print('root_mean_squared_error  igual a ' + str(mean_squared_error (previsao, teste)**0.5))
    print('mean_squared_error  igual a ' + str(mean_squared_error (previsao, teste)))
    print('mean_absolute_error  igual a ' + str(mean_absolute_error (previsao, teste)))
```

```python
#Para todos os mtodos, quanto mais prximo de "Zero", melhor.
desvio_3 (pred_X_test, y_test)
```

```python
#Plotando grfico para visualizar melhor o comportamento do modelo - Laranja valor real, Azul valor previsto pelo modelo
sns.histplot(novo_df_enem_bahia['pred_X_test'], bins = 50, alpha = .5)
sns.histplot(novo_df_enem_bahia['y_test'], color = "orange", bins = 50, alpha = 0.50);
```

```python
# Criando o ambiente do grfico 
sns.set_style("white")
plt.figure(figsize=(10, 10))

# Grfico de Disperso
g = sns.scatterplot(data=novo_df_enem_bahia, alpha = .7)
plt.show()
```
